## Functions in Python

As in most languages, a function can be defined as a block of code that accepts some input, processes them in a certain way and gives us an out.  
Or it can be defined as block of code that performs a certain specific task

```python
# Defining a function in Python
def function_name(variable):
    return variable
    
# Calling the function
function_name('Learn Python')
```

```python
# Defining a more complex function
# This function will take in a list of names (string variables)
# and then return the number of characters in each name
def count_length_of_names(list_names):
    # declare an empty list that will hold the
    # length of each name
    length_list = []
    # Loop through each name
    for name in list_names:
        # Then using the built-in 'len' function get the length of the name
        length_list.append(len(name))
    # return the 'length_list'
    return length_list
```

```python
my_list = ['Kumar Sangakkara', 'Ricky Ponting', 'Sachin Tendulkar', 'Damien Martyn', 'VVS Laxman']

count_length_of_names(my_list)
```

### Matplotlib - Plotting with Python

A nice piece of history. The author of Matplotlib, John Hunter, created this library because back in the early 2000's, Python couldn't match up to the plotting capabilities of R. Some of the syntax and ideas in matplotlib are influenced by R design

```python
%matplotlib inline
```

```python
# Let's import matplotlib
# Just like importing numpy, this is the standard way to do it
import matplotlib.pyplot as plt

# Make an array.
# And here's a new numpy function 'arange'. Similar to range in base Python
array_x = np.arange(0.5, 25.5, 0.75)
# Apply a function to the array (Square it)
array_y = array_x **2

print(array_x)
print(array_y)
```

```python
# Plot the function
plt.plot(array_x, array_y)
```

```python
# Creating subplots

array_z = array_x **4

# 1st subplot
plt.subplot(211)
# What does '211' mean?
# Basically that there are 2 subplots and this one is in Row 1, Column 1
plt.plot(array_x, array_y)

# 2nd subplot
plt.subplot(212)
# This one is in Row 2, Column 1
plt.plot(array_x, array_z)
```

```python
# Creating a sin and cosine plot

array_l = np.arange(-5,5,0.1)
# Using 'sin' function in numpy library
array_sin = np.sin(array_l)
# Similar function for 'cosine'
array_cos = np.cos(array_l)

# 1st subplot
plt.subplot(211)
plt.plot(array_l, array_sin)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sin')

# 2nd subplot
plt.subplot(212)
plt.plot(array_l, array_cos)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('Cos')

# To automatically set padding and optimally arrange the outlines
plt.tight_layout()
```

Standard Weight of Fish is a well reviewed methodology to understand about the expected weight of a certain target species at a certain length.<br>
The equation can be written as <br>
<p style='text-align:center;'><span style='font-size:24px;font-weight:bold;'>W=aL<sup>b</sup></span><br>
Where <b>W</b>=Weight of fish<br>
<b>a</b>=Coefficient of growth<br>
<b>L</b>=Length of fish<br>
<b>b</b>=Exponent of growth<p>
When transformed to natural log, the equation presents itself as<br>
<p style='text-align:center;'><span style='font-size:24px;font-weight:bold;'><i>ln</i> W=b <i>ln</i> L+<i>ln</i> a</span></p>
Which is a linear equation of the form<br>
<p style='text-align:center;'><span style='font-size:24px;font-weight:bold;'>y=mx+c</span></p>

```python
import scipy.stats as stats

#Log male carapace width
ln_male_carapace_width=np.log(male_data['carapace_width'])
#Log male weight
ln_male_weight=np.log(male_data['weight'])
#Obtaining values for slope, intercept and regression value using 'Linear regression'
male_slope, male_intercept, male_r_value, male_p_value, male_std_err = stats.linregress(ln_male_carapace_width,ln_male_weight)
```

```python
plt.figure(figsize=(9,9))
inverted_male_intercept=np.exp(male_intercept)
#The model carapace width value would be set from 50 to 180 mm
male_carapace_width_array=np.arange(50,180,0.1)
#The weight variable
male_weight_result=np.power(male_carapace_width_array,male_slope)*inverted_male_intercept

#Plotting the scatter plot as usual
plt.scatter(male_data['carapace_width'],male_data['weight'],color='#ff8533',alpha=0.75,s=120)
#Plotting the line depicting length weight relationship
plt.plot(male_carapace_width_array,male_weight_result,linestyle='-',color='black',linewidth=0.5)

plt.xlabel('CW\n(mm)')
plt.ylabel('W\n(g)')
plt.ylim(0,550)

#Putting some text in the bottom right of the plot
left_x,right_x=plt.xlim()
bottom_y,top_y=plt.ylim()
plt.text(right_x,bottom_y,'$W=aL^b$\n$a$={0:.7f}\n$b=${1:.2f}'.format(inverted_male_intercept,male_slope),horizontalalignment='right',verticalalignment='bottom',fontsize=20)
plt.text(right_x,top_y,'n={0}'.format(male_data.shape[0]),horizontalalignment='right',verticalalignment='top', fontsize=20)

#Title of the plot
plt.title('Standard Length-Weight Relationship in male Blue Swimming Crab\nin Jaffna fishery, Palk Bay, Sri Lanka',horizontalalignment='center')

plt.tight_layout()
```
